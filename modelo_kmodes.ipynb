# K modes - No supervisado

```python
import pandas as pd
from kmodes.kmodes import KModes
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("intrafamiliar_modelov2.csv")
```

```python
df.head()
```

```python
df_categorico = df.select_dtypes(include=['object', 'category']).copy()
df_categorico.head()
```

```python
le = LabelEncoder()
for col in df_categorico.columns:
    df_categorico[col] = le.fit_transform(df_categorico[col].astype(str))
```

```python
km = KModes(n_clusters=3, init='Huang', n_init=5, verbose=1, random_state=42)
```

```python
clusters = km.fit_predict(df_categorico)
```

```python
df['cluster'] = clusters


print(df['cluster'].value_counts())
print(df.groupby('cluster').mean(numeric_only=True))

```

```python
df.to_csv("intrafamiliar_clusters.csv", index=False)
```

```python
for col in df.select_dtypes(include=['object', 'category']).columns:
    print(f"\n{col} por cluster:")
    print(df.groupby('cluster')[col].agg(lambda x: x.mode()[0]))
```
