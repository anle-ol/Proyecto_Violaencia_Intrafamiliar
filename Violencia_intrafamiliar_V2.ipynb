identificar las variables que influyen en la severidad de la agresión (medida por Días de Incapacidad Medicolegal) y el Factor Desencadenante de la Agresión.

```python
intrafamiliar["Días de Incapacidad Medicolegal"].value_counts().sort_index()
```

```python
#Cerca del 14% del total son Sin Información, no me sirven para mi analisis, y si los reemplazo con
#la Moda, sería de 1  a 30 días, severidad Baja y estaría sesgando los datos a que haya mas de esos.
```

```python
# 1. Definir el mapeo de texto a valor ordinal
mapeo_severidad = {
    'Cero': 0,        # 0 días de incapacidad
    '1 a 30': 1,       # Severidad Baja/Media
    '31 a 90': 2,      # Severidad Media/Alta
    'Más de 90': 3,    # Alta Severidad
    # Los valores 'Sin información' se mapean a NaN para tratarlos como nulos
    'Sin información': np.nan
}


# 2. Aplicar el mapeo para crear la nueva columna 'Severidad_Ordinal'
intrafamiliar['Severidad_Ordinal'] = intrafamiliar['Días de Incapacidad Medicolegal'].replace(mapeo_severidad)

print("--- 1. Conteo de la Nueva Columna con Nulos (antes de excluir) ---")
print(intrafamiliar['Severidad_Ordinal'].value_counts(dropna=False))
```

```python
print(f"\nRegistros totales antes de la exclusión: {len(intrafamiliar)}")

# 3. Eliminar las filas donde 'Severidad_Ordinal' es NaN (los 'Sin información')
intrafamiliar.dropna(subset=['Severidad_Ordinal'], inplace=True)

# 4. Convertir la columna a entero, ya que no tiene nulos y es ordinal
intrafamiliar['Severidad_Ordinal'] = intrafamiliar['Severidad_Ordinal'].astype(int)

print(f"Registros después de excluir 'Sin información': {len(intrafamiliar)}")

print("\n--- 2. Conteo final de la variable objetivo (Sin Nulos) ---")
print(intrafamiliar['Severidad_Ordinal'].value_counts())
```

```python
valor_faltante = 'Sin información'

# 1. Monitoreo inicial
registros_antes = len(intrafamiliar)

# 2. Excluir las filas que tienen el valor faltante en la columna de interés
intrafamiliar = intrafamiliar[
    intrafamiliar['Factor Desencadenante de la Agresión'] != valor_faltante
].copy()

# 3. Monitoreo final
registros_despues = len(intrafamiliar)

print(f"✅ Registros de 'Factor Desencadenante' eliminados: {registros_antes - registros_despues}")
print(f"Registros restantes para el análisis: {registros_despues}")
```

Aplicar One-Hot Encoding (OHE)

```python
# 1. Aplicar pd.get_dummies() para OHE
df_encoded_factor = pd.get_dummies(
    intrafamiliar['Factor Desencadenante de la Agresión'],
    prefix='Factor'
)

# 2. Unir las nuevas columnas codificadas al DataFrame original
# El parámetro axis=1 asegura que se unen como nuevas columnas.
intrafamiliar = pd.concat([intrafamiliar, df_encoded_factor], axis=1)

# 3. Eliminar la columna original de texto
intrafamiliar.drop('Factor Desencadenante de la Agresión', axis=1, inplace=True)

print("\n--- Columnas OHE del Factor Desencadenante Creadas ---")
# Muestra las nuevas columnas binarias
print(intrafamiliar.filter(like='Factor_').head())
```

Limpieza y OHE de Presunto Agresor Detallado

```python
intrafamiliar["Presunto Agresor Detallado"].value_counts()
```

```python
# Verificamos si es buena idea agrupar por Familiar Directo
#Si Padre es muy diferente de Madre, entonces NO se agrupa, si tiene valores similares, se agrupan

categorias_directas = ['Padre', 'Madre', 'Hijo (a)', 'Hermano (a)']

# 1. Filtra solo las filas de este grupo
df_filtro = intrafamiliar[intrafamiliar['Presunto Agresor Detallado'].isin(categorias_directas)]

# 2. Calcula la severidad media para cada subcategoría
severidad_por_familiar = df_filtro.groupby('Presunto Agresor Detallado')['Severidad_Ordinal'].mean().sort_values(ascending=False)

print("--- Severidad Media de Agresores Familiares Directos ---")
print(severidad_por_familiar)
```

```python
#El valor 0.8 significa que el caso promedio de agresión dentro de este grupo está más cerca de la categoría 1
#(1 a 30 dúas de incapacidad) que de la categoría 0 (Cero días), pero aún es una severidad muy baja en la escala general (donde el máximo es 3).

#Dado que las medias de severidad para el Padre, Madre, Hijo(a) y Hermano(a) son todas extremadamente cercanas
#y se encuentran en el mismo rango de riesgo, la pérdida de información por agruparlas es mínima y el beneficio
#en la simplificación del modelo es máximo.
```
